ELECTRICAL CIRCUIT
LR ckt.

The differential equation of LR ckt is L\frac{di}{dt}+Ri=E

\frac{di}{dt}+\frac{R}{L}i=\frac{E}{L}

Problem: A coil which has an inductance of 40 h and a resistance of 2Ω is connected together to form a LR series circuit. If they are connected
to a 20V DC supply.What will be the current flowing through the ckt.If initialy at time t=0 there is no current flowing through the ckt.

given L=40 ,R=2 & E=20, i(0)=0

```python
var('t')
i=function('i')(t)
i=desolve(diff(i,t)+2/40*i-20/40,i,[0,0])
show(i)
```

```python
plot(i)
```
