4.Find the factors of sum of digits of 1275!

```python
n=factorial(1275)
t_sum=sum(int(d) for d in str(n))
t_sum
factors=[]
for i in range(1,t_sum+1):
    if t_sum % i==0:
        factors.append(i)
factors
```
